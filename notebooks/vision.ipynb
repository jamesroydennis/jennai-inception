## Executive Summary
In a global economy undergoing rapid shifts ‚Äî from tariffs to currency volatility ‚Äî opportunities emerge for agile entrepreneurs to exploit inefficiencies in pricing and logistics. While basic necessities such as food, energy, and water are foundational to life, these markets are highly regulated, capital-intensive, and difficult to penetrate. This project instead identifies **textiles** as a leaner, scalable entry point.

## Why Textiles?
- **Resilient Demand:** Clothing remains a necessity with consistent consumer demand.
- **Global Trade Exposure:** Textile costs fluctuate sharply due to trade agreements, tariffs, and FX swings.
- **Favorable Margins:** Lightweight, shippable, high markup potential.
- **Low Regulatory Barrier:** Compared to energy or food, textile import/export is easier to navigate.

By targeting T-shirts as a test case, we follow the "dental floss" strategy: start small, string across the river, and build a bridge.

## Project Purpose
To create an intelligent sourcing system that automatically scores the best global region to source a textile product from, based on:

- **Tariffs & trade policy**
- **Currency conversion rates & volatility**
- **Supply chain availability and delays**
- **Unpredictability Quotient (UQ)**: a combined risk signal for instability

This enables any seller or distributor to make better daily purchasing decisions ‚Äî like day trading for physical goods.
